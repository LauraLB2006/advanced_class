```python
import pandas as pd
```

```python
eq_csv = pd.read_csv("all_day.CSV")
eq_csv.head()
```

```python
new_eq= eq_csv.drop (columns=['nst', 'gap', 'dmin', 'rms', 'net'])
```

```python
new_eq
```

```python
locations = new_eq.iloc [:,1:3].values.tolist()
```

```python
locations
```

```python
import arcgis
from arcgis.features import SpatialDataFrame

sdf = SpatialDataFrame(new_eq,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```

```python
sdf1 = sdf[sdf['mag'] < 2.5]
```

```python
sdf2 = sdf[(sdf['mag']>2.5) & (sdf['mag'] < 4)]
```

```python
sdf3 = sdf[(sdf['mag']>4) & (sdf['mag'] < 5)]
```

```python
sdf3
```

```python
sdf4 = sdf[sdf['mag'] > 5]
```

```python
sdf4
```

```python
sdf3.reset_index(inplace=True, drop=True)
```

```python
from arcgis.gis import GIS
from IPython.display import display

gis = GIS("http://slustl.maps.arcgis.com/home", "lbernett_SLUSTL")

```

```python
eq = gis.content.import_data(sdf3, title='Magnitude Between 4 and 5')
```

```python
eq_map = gis.map('USA', zoomlevel=4)
eq_map
```

```python
eq_map.add_layer(eq, {"renderer":"HeatmapRenderer",
               "opacity": 0.75})
```
