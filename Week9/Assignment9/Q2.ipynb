```python
import pandas as pd
```

```python
df1 = pd.read_csv("all_week.CSV")
df1.head()
```

```python
df2 = pd.read_csv("all_week_aug_13_20.csv")
df2.head()
```

```python
comb_df = df1.append(ignore_index=True, other=df2)
```

```python
locations = comb_df.iloc[:,1:3].values.tolist()
```

```python
import arcgis
from arcgis.features import SpatialDataFrame

sdf = SpatialDataFrame(comb_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```
