```python
import requests
import pandas as pd
import json

yr = "2015"
m = "1"
c = "FRA"

url = pd.read_json (r'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year='+(yr)+'&month='+(m)+'&coa='+ (c))

#url = 'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=2015&month=1&coa=FRA'

data = requests.get(url)
as_json = data.json()
df = pd.DataFrame(as_json)
df.sort_values (by= "value")
df
```
